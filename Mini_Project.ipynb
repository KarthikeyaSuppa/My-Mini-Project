## size and category

```python
# prompt: convert size column into numeric as per the above cell

def convert_size(size):
  if size == 'S':
    return 1
  elif size == 'M':
    return 2
  elif size == 'L':
    return 3
  elif size == 'XL':
    return 4
  else:
    return 5

data['size'] = data['size'].apply(convert_size)
data.head()
```

```python
# prompt: convert size into numeric

category_mapping = {
    'Classic': 1,
    'Supreme': 2,
    'Veggie': 3,
    'Chicken': 4,
    }
data['category'] = data['category'].map(category_mapping)
```

```python
data.describe()
```

## pizza_type_id

```python
# prompt: unique of pizza_type_id

unique_pizza_types = data['pizza_type_id'].unique()

print(f"Unique pizza types: {unique_pizza_types}")
```

```python
# prompt: generate code for changing above cell unique values into numeric for the column

pizza_type_mapping = {}
for i, pizza_type in enumerate(unique_pizza_types):
  pizza_type_mapping[pizza_type] = i + 1

data['pizza_type_id'] = data['pizza_type_id'].map(pizza_type_mapping)
data.head()
```

## ingredients

```python
# prompt: unique of ingreient1,2,3,4,5,6,7,8

# Get unique values for each ingredient column
unique_ingredients = {}
for i in range(1, 9):
  unique_ingredients[f"ingredient_{i}"] = data["ingredient" + str(i)].unique()

# Print the unique values for each ingredient column
for ingredient, unique_values in unique_ingredients.items():
  print(unique_values)
```

```python
# prompt: merge all the unique values itn single list

# Extract all unique values from the ingredient columns
unique_values = []
for ingredient, unique_list in unique_ingredients.items():
  unique_values.extend(unique_list)

# Remove duplicates from the list
unique_values = list(set(unique_values))

# Print the merged list
print(unique_values)
```

```python
# prompt: generate code for changing above cell unique values into numeric for the column ingredient1,2,3,4,5,6,7,8

ingredient_mapping = {}
for i, ingredient in enumerate(unique_values):
  ingredient_mapping[ingredient] = i + 1

for i in range(1, 9):
  data[f"ingredient{i}"] = data[f"ingredient{i}"].map(ingredient_mapping)

data.head()
```
